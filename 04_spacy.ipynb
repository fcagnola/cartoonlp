```python
import pandas as pd
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob
from typing import List
```

```python
df = pd.read_pickle('03_out_dataframe.pickle')
```

```python
df.iloc[0]['Sentence_Tokenized']
```

```python
def sentiment(text: List[str]):
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe('spacytextblob')
    polarities = []
    for sentence in text:
        doc = nlp(sentence)
        polarities.append(doc._.polarity)
    return polarities
```

```python
df['Sentence_Sentiment'] = df['Sentence_Tokenized'].apply(lambda x: sentiment(x))
```

```python
df.to_pickle('04_out_sentiment.pickle')
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_sentiment(title, sentiment):
    y_points = np.array(sentiment)
    x_points = np.array(range(len(sentiment)))
    plt.figure()
    plt.plot(x_points, y_points)
    plt.title(title)
    plt.xlabel("Movie duration")
    plt.ylabel("Sentiment score")
    plt.savefig(f"spacy_sentiment/{title}_sentiment_spacy.jpg")
    plt.close()
```

```python
for title, sentiment in df['Sentence_Sentiment'].iteritems():
    plot_sentiment(title, sentiment)
```
