```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
```

```python
df = pd.read_pickle('04_out_sentiment.pickle')

df.head(59)
```

```python
fdist = nltk.FreqDist(df['Tokenized'].iloc[0])
```

```python
filtered = [(m, n) for m, n in fdist.items() if n > 3 and m not in stopwords.words('english')]
```

```python
for key, val in sorted(filtered, key=lambda x: x[1], reverse=True):
    print(key, val)
```
